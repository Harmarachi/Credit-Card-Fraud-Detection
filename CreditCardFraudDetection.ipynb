### Scaling our data

Though the anonymized features have been scaled, our time and amount features have not. If we do not scale them too, it could result in giving weights to features and therefore poor performance. To avoid this issue, we should standardize both the time and amount columns.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# scaling time
scaled_time = scaler.fit_transform(data[['Time']])
flat_list1 = [item for sublist in scaled_time.tolist() for item in sublist]
scaled_time = pd.Series(flat_list1)

# scaling amount
scaled_amount = scaler.fit_transform(data[['Amount']])
flat_list2 = [item for sublist in scaled_amount.tolist() for item in sublist]
scaled_amount = pd.Series(flat_list2)
```

```python
# Adding the new columns in the original dataframe
scaled_data = pd.concat([data, scaled_amount.rename('scaled_amount'), scaled_time.rename('scaled_time')], axis=1)

# Dropping the old time and amount columns
scaled_data.drop(['Amount', 'Time'], axis=1, inplace=True)
scaled_data
```

### Declare the dependent and independent variables

```python
x = scaled_data.iloc[:,:-1]
y = scaled_data['Class']
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
classifier = linear_model.LogisticRegression(C=1e5)
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = np.array(classifier.predict(x_test))
y = np.array(y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
